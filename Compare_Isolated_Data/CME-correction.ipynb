```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from matplotlib import cm
```

```python
maxImages = 20

fig, axs = plt.subplots(1,1, figsize=(7,7))
axs.plot(1,0, 'yo', markersize=30)

posPositionArray = np.load("output_compare_data/posPositionArray.npy")
negPositionArray = np.load("output_compare_data/negPositionArray.npy")
posPositionArrayAdj = np.load("output_compare_data/posPositionArrayAdj.npy")
negPositionArrayAdj = np.load("output_compare_data/negPositionArrayAdj.npy")

posPositionArrayIso = np.load("output_compare_data/posPositionArrayIso.npy")
negPositionArrayIso = np.load("output_compare_data/negPositionArrayIso.npy")
posPositionArrayAdjIso = np.load("output_compare_data/posPositionArrayAdjIso.npy")
negPositionArrayAdjIso = np.load("output_compare_data/negPositionArrayAdjIso.npy")

posPositionArrayNosub = np.load("output_compare_data/posPositionArrayNosub.npy")
negPositionArrayNosub = np.load("output_compare_data/negPositionArrayNosub.npy")
posPositionArrayAdjNosub = np.load("output_compare_data/posPositionArrayAdjNosub.npy")
negPositionArrayAdjNosub = np.load("output_compare_data/negPositionArrayAdjNosub.npy")


axs.plot(posPositionArray[:maxImages+1,0],posPositionArray[:maxImages+1,1],color='blue', label='Subtracted Data, Positive Sqrt ', linestyle='solid')
axs.plot(negPositionArray[:maxImages+1,0],negPositionArray[:maxImages+1,1],color='orange', label='Subtracted Data, Negative Sqrt ', linestyle='solid')
axs.plot(posPositionArrayAdj[:maxImages+1,0],posPositionArrayAdj[:maxImages+1,1],color='blue', label='Corrected Positive Position', linestyle='solid')
axs.plot(negPositionArrayAdj[:maxImages+1,0],negPositionArrayAdj[:maxImages+1,1],color='orange', label='Corrected Negative Position', linestyle='solid')

axs.plot(posPositionArray[maxImages:,0],posPositionArray[maxImages:,1],color='blue', linestyle='solid')
axs.plot(negPositionArray[maxImages:,0],negPositionArray[maxImages:,1],color='orange', linestyle='solid')
axs.plot(posPositionArrayAdj[maxImages:,0],posPositionArrayAdj[maxImages:,1],color='lightsteelblue', linestyle='solid')
axs.plot(negPositionArrayAdj[maxImages:,0],negPositionArrayAdj[maxImages:,1],color='navajowhite', linestyle='solid')

axs.plot(posPositionArrayIso[:maxImages+1,0],posPositionArrayIso[:maxImages+1,1],color='green', label='Isolated Data, Positive Sqrt ', linestyle='solid')
axs.plot(negPositionArrayIso[:maxImages+1,0],negPositionArrayIso[:maxImages+1,1],color='red', label='Isolated Data, Negative Sqrt ', linestyle='solid')
axs.plot(posPositionArrayAdjIso[:maxImages+1,0],posPositionArrayAdjIso[:maxImages+1,1],color='green', label='Corrected Positive Position Isolated', linestyle='solid')
axs.plot(negPositionArrayAdjIso[:maxImages+1,0],negPositionArrayAdjIso[:maxImages+1,1],color='red', label='Corrected Negative Position Isolated', linestyle='solid')

axs.plot(posPositionArrayIso[maxImages:,0],posPositionArrayIso[maxImages:,1],color='green', linestyle='solid')
axs.plot(negPositionArrayIso[maxImages:,0],negPositionArrayIso[maxImages:,1],color='red', linestyle='solid')
axs.plot(posPositionArrayAdjIso[maxImages:,0],posPositionArrayAdjIso[maxImages:,1],color='palegreen', linestyle='solid')
axs.plot(negPositionArrayAdjIso[maxImages:,0],negPositionArrayAdjIso[maxImages:,1],color='lightcoral', linestyle='solid')

axs.plot(posPositionArrayNosub[:maxImages+1,0],posPositionArrayNosub[:maxImages+1,1],color='black', linestyle='solid')
axs.plot(negPositionArrayNosub[:maxImages+1,0],negPositionArrayNosub[:maxImages+1,1],color='black', linestyle='solid')
axs.plot(posPositionArrayAdjNosub[:maxImages+1,0],posPositionArrayAdjNosub[:maxImages+1,1],color='black', linestyle='solid')
axs.plot(negPositionArrayAdjNosub[:maxImages+1,0],negPositionArrayAdjNosub[:maxImages+1,1],color='black', linestyle='solid')

axs.plot(posPositionArrayNosub[maxImages:,0],posPositionArrayNosub[maxImages:,1],color='black', label='No subtraction ',linestyle='solid')
axs.plot(negPositionArrayNosub[maxImages:,0],negPositionArrayNosub[maxImages:,1],color='black',linestyle='solid')
axs.plot(posPositionArrayAdjNosub[maxImages:,0],posPositionArrayAdjNosub[maxImages:,1],color='black', linestyle='solid')
axs.plot(negPositionArrayAdjNosub[maxImages:,0],negPositionArrayAdjNosub[maxImages:,1],color='black', linestyle='solid')


thomson = plt.Circle((0.5, 0), 0.5, color='k', fill=False)
axs.add_patch(thomson)
axs.set_aspect('equal')
axs.set_ylim(0,0.6)
axs.set_xlim(0,1.1)


axs.set_xlabel('x (au)')
axs.set_ylabel('y (au)')

axs.legend(loc='upper left')

plt.savefig('compare_subtraction.png', dpi=300)

plt.show()
```
